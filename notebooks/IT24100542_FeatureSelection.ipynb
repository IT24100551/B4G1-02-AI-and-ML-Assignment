```python
# Feature Engineering

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file into a DataFrame named 'df'
df = pd.read_csv('/content/winequality-red.csv')

# Create new feature
df['sulphates_alcohol_ratio'] = df['sulphates'] / df['alcohol']

df[['sulphates','alcohol','sulphates_alcohol_ratio']].head()


plt.scatter(df['alcohol'], df['sulphates_alcohol_ratio'], alpha=0.5)
plt.xlabel("Alcohol")
plt.ylabel("Sulphates/Alcohol Ratio")
plt.title("New Feature: Sulphates to Alcohol Ratio")
plt.show()
```
